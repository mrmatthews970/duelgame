```python
class Hero():
    def __init__(self,name):
        self.name = name
        
class Swordsman(Hero):
    def __init__(self,name):
        super().__init__(name)
        self.desc = "Swordsman"
        self.health = 100
        self.armor = 50
        self.attack = 75
        self.endurance = 5
        self.weapon = "Sword"
    def strike(self,enemy):
        if self.endurance > 0 and self.health > 0:
            enemy.armor -= self.attack
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 1
    def speed_attack(self,enemy):
        if self.endurance >= 5 and self.health > 0:
            enemy.armor -= self.attack*3
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 5
    def rest(self):
        self.endurance = 5
    def miss(self):
        self.endurance -= 1
        
class Magician(Hero):
    def __init__(self,name):
        super().__init__(name)
        self.desc = "Magician"
        self.health = 75
        self.armor = 25
        self.attack = 50
        self.endurance = 10
        self.weapon = "Magic"
    def strike(self,enemy):
        if self.endurance > 0 and self.health > 0:
            enemy.armor -= self.attack
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 1
    def rest(self):
        self.endurance = min(10,self.endurance + 5)
    def heal(self):
        if self.health > 0:
            self.health = min(75,self.health + 10)
        self.endurance -= 2
    def miss(self):
        self.endurance -= 1
            
class Brute(Hero):
    def __init__(self,name):
        super().__init__(name)
        self.desc = "Brute"
        self.health = 125
        self.armor = 100
        self.attack = 100
        self.endurance = 5
        self.weapon = "Hammer"
    def strike(self,enemy):
        if self.endurance > 0 and self.health > 0:
            enemy.armor -= self.attack
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 1
    def resistance(self):
        if self.endurance == 5 and self.health > 0:
            self.armor += 150
        self.endurance -= 5
    def rest(self):
        self.endurance = 5
    def miss(self):
        self.endurance -= 1
            
class Horseman(Hero):
    def __init__(self,name):
        super().__init__(name)
        self.desc = "Horseman"
        self.health = 75
        self.armor = 50
        self.attack = 50
        self.endurance = 10
        self.weapon = "Spear"
    def strike(self,enemy):
        if self.endurance > 0 and self.health > 0:
            enemy.armor -= self.attack
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 1
    def charge(self,enemy):
        if self.endurance >= 5 and self.health > 0:
            enemy.armor -= self.attack*2
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 5
    def rest(self):
        self.endurance = min(10,self.endurance + 5)
    def miss(self):
        self.endurance -= 1
            
class Healer(Hero):
    def __init__(self,name):
        super().__init__(name)
        self.desc = "Healer"
        self.health = 100
        self.armor = 50
        self.attack = 25
        self.endurance = 10
        self.weapon = "Staff"
    def strike(self,enemy):
        if self.endurance > 0 and self.health > 0:
            enemy.armor -= self.attack
            if enemy.armor <= 0:
                enemy.health += enemy.armor
                enemy.armor = 0
            if enemy.health <= 0:
                enemy.health = 0
            self.endurance -= 1
    def rest(self):
        self.endurance = min(10,self.endurance + 5)
    def self_heal(self):
        if self.health < 75 and self.health > 0:
            self.health = min(75,self.health + 25)
        if self.armor < 50:
            self.armor = min(50,self.armor + 25)
        self.endurance -= 1
    def team_heal(self,teammate):
        if teammate.desc == "Swordsman":
            teammate.health = min(100,teammate.health + 25)
            teammate.armor = min(50,teammate.armor + 25)
            teammate.endurance = min(5,teammate.endurance + 1)
        elif teammate.desc == "Magician":
            teammate.health = min(75,teammate.health + 25)
            teammate.armor = 25
            teammate.endurance = min(10,teammate.endurance + 1)
        elif teammate.desc == "Brute":
            teammate.health = min(125,teammate.health + 25)
            teammate.armor = min(100,teammate.armor + 25)
            teammate.endurance = min(5,teammate.endurance + 1)
        elif teammate.desc == "Horseman":
            teammate.health = min(75,teammate.health + 25)
            teammate.armor = min(50,teammate.health + 25)
            teammate.endurance = min(10,teammate.endurance + 1)
    def miss(self):
        self.endurance -= 2
```
